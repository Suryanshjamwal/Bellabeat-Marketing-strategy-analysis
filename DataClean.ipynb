## Checking for duplicates

```python
len(dfAct)-len(dfAct.drop_duplicates())
```

```python
len(dfSleep)-len(dfSleep.drop_duplicates())
```

```python
len(dfHStep)-len(dfHStep.drop_duplicates())
```

```python
# dropping duplicates in Sleep dataset

dfS = dfSleep.drop_duplicates()
len(dfS)-len(dfS.drop_duplicates())

# dfS is the updated Sleep dataset
```

## Converting timestamps

```python
dfAct['ActivityDate'] = pd.to_datetime(dfAct['ActivityDate'], format='%m/%d/%Y')
dfAct['day_of_week'] = dfAct['ActivityDate'].dt.day_name()
```

```python
dfAct
```

```python
dfS[['Date', 'Time']] = dfS['SleepDay'].str.split(' ', 1, expand=True)
```

```python
dfS['Date'] = pd.to_datetime(dfS['Date'], format='%m/%d/%Y')
dfS['day_of_week'] = dfS['Date'].dt.day_name()
```

```python
dfHStep[['Date', 'Time', 'AM/PM']] = dfHStep['ActivityHour'].str.split(' ', 2, expand=True)
```

```python
dfHStep['Hour'] = dfHStep['Time'] + " " + dfHStep['AM/PM']
dfHStep
```

```python
dfHStep['Date'] = pd.to_datetime(dfHStep['Date'], format='%m/%d/%Y')
dfHStep['day_of_week'] = dfHStep['Date'].dt.day_name()
```

## Data Merging

```python
dfActS = pd.merge(dfAct, dfS,  how='left', left_on=['Id','ActivityDate'], right_on = ['Id','Date'])
```

```python
dfActS
```

```python
dfActS.describe()
```

### Key Takeaways:
    * Average sleep hours is 7 Hours.
